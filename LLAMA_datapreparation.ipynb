```python
import io
from sklearn.model_selection import KFold
import numpy as np
from transformers import AutoModelForCausalLM, AutoTokenizer
from huggingface_hub import notebook_login

k=4
seed = 1
kf = KFold(n_splits=k,random_state=seed, shuffle=True)
#notebook_login()
for i, (train_index, test_index) in enumerate(kf.split(LLM_data)):
  train,test = LLM_data['text'].iloc[train_index], LLM_data['text'].iloc[test_index]

  # print(np.shape(test))
  # print(np.shape(train))
  test_data = pd.DataFrame(test)
  #print(test_final['text'].str.split('###Output: ').str)
  tes = test_data['text'].str.split('###Output: ').str

  test_final = tes[0] + '###Output: '
  test_new_final = pd.DataFrame(test_final)
  #print(test_new_final)
  test_new_final.to_csv('/content/data/test_'+str(i)+'.csv')
  train.to_csv('/content/data/train_'+str(i)+'.csv')
  test_new_final.to_csv('/content/data/dev_'+str(i)+'.csv')

#   run_llama()
  !autotrain llm --train --project_name 'llama2-causalityextractionCV' --model yahma/llama-7b-hf --data_path /content/data/ --use_peft --use_int4 --learning_rate 2e-4 --train_batch_size 2 --num_train_epochs 3 --push_to_hub --repo_id Seetha/llama2-causalityextractioncv --token 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 --trainer sft # > training.log &

  tokenizer = AutoTokenizer.from_pretrained("Seetha/llama2-causalityextractioncv")
  model = AutoModelForCausalLM.from_pretrained("Seetha/llama2-causalityextractioncv")

  input_ids_sents = []
  for i in test_new_final['text']:
    input_ids = tokenizer.encode(i, return_tensors="pt")
    print(input_ids)
    input_ids_sents.append(input_ids)

  for j in input_ids_sents:
  #print(j)
    output = model.generate(
        input_ids=j, max_length=256, temperature=0.0, num_return_sequences=1
    )  #num_return_sequences=1
    #print(tokenizer.decode(output[0]))
    classification_prediction_list =[]
    classification_prediction_list.append(tokenizer.decode(output[0]))

    test_new_final.loc[i, "LLAMA_Predictions"]=classification_prediction_list
test_new_final.to_excel('/content/LLAMA_GD_predictions.xlsx')
```
