```python
from transformers import AutoModelForCausalLM, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("Seetha/llama2-causalityextractionzero-sot")
model = AutoModelForCausalLM.from_pretrained("Seetha/llama2-causalityextractionzero-sot")
```

```python
data_test
```

```python
input_ids_sents = []
for i in data_test['transfer']:
    input_ids = tokenizer.encode(i, return_tensors="pt")
    #print(input_ids)
    input_ids_sents.append(input_ids)
predictions_transfer = []
for j in input_ids_sents:
  #print(j)
  output = model.generate(
        input_ids=j, max_length=256, temperature=0.0, num_return_sequences=1
  )  #num_return_sequences=1
  print(tokenizer.decode(output[0]))
  predictions_transfer.append(tokenizer.decode(output[0]))

preds = pd.DataFrame(predictions_transfer)
preds.to_excel('/content/LLAMA-preds_zero-shot5epochs.xlsx')
```
